```python
#importing Pandas
import pandas as pd
#importing Decision Tree regressor from scikit-learn
from sklearn.tree import DecisionTreeRegressor
#importing mean_absolute_error
from sklearn.metrics import mean_absolute_error
#importing train_test_split
from sklearn.model_selection import train_test_split
```

```python
#importing our data that is in csv format and showing a sample of it
#data is taken from "https://www.kaggle.com/datasets/dansbecker/melbourne-housing-snapshot"
melbourne_Data_path = "/content/melb_data.csv"
melbourneData = pd.read_csv(melbourne_Data_path)
melbourneData.describe()
```

```python
#showing the columns our data contains, this will be important in choosing the features we need
melbourneData.columns
```

```python
#dropping the missing data
melbourneData.dropna(axis=0)
```

```python
#choosing our prediction target, the variable that we will predict
y = melbourneData.Price
```

```python
#choosing features we will train our model with and storing them in a dataframe called "x"
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude'] #list of the features we need
x = melbourneData[melbourne_features] #dataFrame of the features we need
x.describe()
```

```python
#shows the first rows of the data 
x.head()
```

```python
#splitting the data into testing data and validation data
train_X, val_X, train_y, val_y = train_test_split(x, y,random_state=1)
```

```python
#Defining the model's type and setting the random_state to a number means we will have the same result each time
melbourneModel = DecisionTreeRegressor(random_state=1)
#Fitting the model, training the model on the training data we have (x) to predict the variable we want (y)
#in this case we fit our model using the training data part that we split previously
melbourneModel.fit(train_X, train_y)
```

```python
val_X.dropna(axis=0)
```

```python
# we get the predictions here
val_predictions = melbourneModel.predict(val_X)
```

```python
# Checking the predictions against the validations data
print(val_predictions[:5])
print("-------------")
print(val_y[:5])
```
